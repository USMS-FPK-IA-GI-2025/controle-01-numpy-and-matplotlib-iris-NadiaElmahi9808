# Partie A — Importation du Dataset, Arrays, Premières Manipulations

### A1. Importation du dataset
Importez NumPy et Matplotlib.  
Chargez le fichier `iris.csv` avec `np.genfromtxt` (`delimiter=','`, `skip_header=1`).  
Stockez les données dans un array nommé `data` et affichez sa `shape`.


```python
import numpy as np
import matplotlib.pyplot as plt
#Charger le fichier
data = np.genfromtxt('data/Iris.csv', delimiter=',', dtype=str,skip_header=1)
print(f"Dimension de notre dataset :{data.shape}")
```

### A2. Séparation des features et des espèces
Séparez :
- les 4 premières colonnes numériques dans un array `features`
- la dernière colonne (espèces) dans un array `species`
Utilisez le slicing.


```python
features = data[:,1:5].astype(float) #toutes les lignes et 1--> 4 colonne car 1ere colonne c'est l'ID
species = data[:,-1] # Toutes les lignes et la 4eme colonnes
#verification
print(f"Dimension de features : {features.shape}")
print(f"Dimension de species : {species.shape}")
```

### A3. Indexing
Affichez les 5 premières lignes de `features` en utilisant l’indexing.


```python
#Lignes 0 --> 4
print(features[:5])
```

### A4. Extraction d’une colonne
Créez un array `sepal_length` contenant uniquement la première colonne de `features`.  
Affichez sa `size` et son `dtype`.


```python
sepal_length = features[:, 0]   #1ere colonne index 0
print(f"Num total des elements dans sepal length : {sepal_length.size}")
print(f"Type de données de sepal length : {sepal_length.size}")
```

### A5. Slicing par espèce
Utilisez le slicing pour extraire les 50 premières lignes correspondant à l’espèce *Setosa*.  
Stockez-les dans `setosa_features`.


```python
#Slicing 50 lignes
setosa_features = features[:50]
#Verification
print(f"Dimension de setosa_features : {setosa_features.shape}")
print(f"N elements dans setosa_features : {setosa_features.size}")
```

# Partie B — Statistiques Basiques et Opérations Mathématiques

### B6. Moyenne par colonne
Calculez la moyenne de chaque colonne de `features` avec `np.mean(axis=0)`.


```python
moyennes_col = np.mean(features, axis=0)
print(moyennes_col)
```

### B7. Min et Max
Trouvez la valeur minimale et maximale de la longueur des pétales.


```python
#longueur des pétales 3eme colonne dans features
petal_length=features[:,2]
min_petal=np.min(petal_length)
max_petal=np.max(petal_length)
print(f"Valeur minimale de la longueur des pétales : {min_petal}")
print(f"Valeur maximale de la longueur des pétales : {max_petal}")
```

### B9. Normalisation
Normalisez les longueurs des pétales en les divisant par leur maximum.
Stockez le résultat dans `normalized_petals`.


```python
normalized_petals = petal_length / max_petal
#Afficher 5 premieres lignes
print(normalized_petals[:5])
```

### B10. Somme cumulée
Calculez la somme cumulée des largeurs des pétales pour Versicolor.
Affichez les 10 premières valeurs.


```python
# Sepal_Width se trouve dans 2eme colonne de features
somme_cumulee = np.cumsum(features[50:100, 1])
print(somme_cumulee[:10])
```

# Partie C — Transformations Avancées avec NumPy

### C11. Espèces uniques
Listez les espèces uniques avec `np.unique` et comptez leurs occurrences.


```python
esp_uniques,n_occ=np.unique(species,return_counts=True)
print(f"Les especes uniques sont :{esp_uniques}")
print(f"Nbre d'occurences est :{n_occ}")
```

### C12. Reshape par espèce
Transformez `features` en un array de forme `(3, 50, 4)` nommé `features_by_species`.


```python
#shape initial (150,4),150 = 30 x 5
features_by_species=features.reshape(3,50,4)
print(f"Dimension des features_by_species : {features_by_species.shape}")
```

### C13. Concaténation
Concaténez horizontalement les moyennes par colonne avec un array de zéros.


```python
zeros=np.zeros(moyennes_col.shape)
concat = np.hstack((moyennes_col,zeros))
print("Resultat de concatenation :")
concat
```

### C15. Broadcasting
Ajoutez 1 à toutes les valeurs de la première colonne de `features`.

### C16. np.where
Créez un array binaire indiquant si la longueur des pétales est supérieure à 4.0.

# Partie E — Visualisations Avancées

### E22. Subplots comparatifs
Deux scatter plots côte à côte : Setosa vs Versicolor.


```python
fig=plt.figure(figsize=(10,5))
sc1=fig.add_subplot(1,2,1)
sc2=fig.add_subplot(1,2,2)
#Setosa
setosa_array=features_by_species[0]
sc1.scatter(setosa_array[:,2],setosa_array[:,3],color='green')
sc1.set_title("Setosa : Largeur vs Longueur petales")
sc1.set_xlabel("Longueur")
sc1.set_ylabel("Largeur")
#Versicolor
ver_array=features_by_species[1]
sc2.scatter(ver_array[:,2],ver_array[:,3],color='blue')
sc2.set_title("Versicolor : Largeur vs Longueur petales")
sc2.set_xlabel("Longueur")
sc2.set_ylabel("Largeur")
plt.savefig("figures/e22.png")
plt.show()
```

### E23. Grille 2x2
Histogrammes des 4 caractéristiques avec `plt.subplots(2,2)`.


```python
fig,axs=plt.subplots(2,2,figsize=(10,10))
axs[0,0].hist(features[:,0],bins=10)
axs[0,0].set_title("Longueur sepales")
axs[0,1].hist(features[:,1],bins=10)
axs[0,1].set_title("Largeur sepales")
axs[1,0].hist(features[:,2],bins=10)
axs[1,0].set_title("Longueur pétales")
axs[1,1].hist(features[:,3],bins=10)
axs[1,1].set_title("Largeur pétales")
plt.savefig("figures/e23.png")
plt.show()
```

### E24. Bar chart avec error bars
Moyennes par espèce avec écart-type.


```python
moy_esp=np.mean(features_by_species,axis=1)
ecart_type=np.std(features_by_species,axis=1)
plt.bar(esp_uniques,moy_esp[:,1],yerr=ecart_type[:,1])
plt.title("Moyennes par espèce avec error bars")
plt.xlabel("Especes")
plt.ylabel("Moyennes")
plt.savefig("figures/e24.png")
plt.show()
```

### E25. Scatter coloré par espèce
Scatter plot avec couleurs par espèce.


```python
fig,ax=plt.subplots(figsize=(10,5))
ax.scatter(setosa_array[:,0],setosa_array[:,1],color='green')
ax.scatter(ver_array[:,0],ver_array[:,1],color='blue')
#Virginica
vir_array=features_by_species[2]
ax.scatter(vir_array[:,0],vir_array[:,1],color='red')
ax.set_title("Largeur vs Longueur sepales")
ax.set_xlabel("Longueur")
ax.set_ylabel("Largeur")
plt.legend(esp_uniques)
plt.savefig("figures/e25.png")
plt.show()
```

### E26. Line plot superposé
Cumsum des largeurs des pétales par espèce.


```python
plt.plot(np.cumsum(setosa_array[:,3]),color='green',label='Setosa')
plt.plot(np.cumsum(ver_array[:,3]),color='blue',label='Versicolor')
plt.plot(np.cumsum(vir_array[:,3]),color='red',label='Virginica')
plt.title("Cumsum des largeurs des pétales par espèce")
plt.xlabel("Entrée")
plt.ylabel("Cumsum")
plt.legend()
plt.savefig("figures/e26.png")
plt.show()
```
